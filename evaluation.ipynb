```python
file_path = "outputs/2025-10-02_14-45-50/result_semantic_2025-10-02_144609.csv"
test_csv = pd.read_csv(file_path)
test_csv
```

```python
with open("outputs/results.md", "w", encoding="utf-8") as f:
    for i, row in test_csv.head().iterrows():
        f.write(f"# Question {i+1}\n\n")
        f.write(f"## User Question\n{row['user_input']}\n\n")
        f.write(f"## Generated Answer\n{row['response']}\n\n")
        f.write(f"## Ground Truth Answer\n{row['reference']}\n\n")
        f.write(f"## Retrieved Chunks\n")
        if isinstance(row["retrieved_contexts"], list):
            for chunk in row["retrieved_contexts"]:
                f.write(f"- {chunk}\n")
        else: 
            f.write(f"{row['retrieved_contexts']}\n")
        f.write("\n---\n\n")
```
